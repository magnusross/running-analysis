```python
import pandas as pd

df = pd.read_parquet("Parkrun_2025.parquet")
```

```python
df = df[df["Time"].str.contains(":")]
```

```python
# Convert the "Time" column to total minutes as a decimal, handling missing values
df["Total Minutes"] = df["Time"].apply(
    lambda x: int(x.split(":")[0]) + int(x.split(":")[1]) / 60
    if pd.notna(x) and ":" in x
    else None
)
```
